```python
def execute(agent, algorithm, policy, hyper_params, penalty=False, cutoff_time=100):
    # Se declara una semilla aleatoria
    random_state = np.random.RandomState(47)

    agent.set_hyper_parameters(hyper_params)
    agent.random_state = random_state
    """
    Declaramos como True la variable de mostrar video,
    para ver en tiempo real cómo aprende el agente.
    Borrar esta línea para acelerar la velocidad del aprendizaje.
    """
    agent.display_video = False #True
    agent.set_cutoff_time(cutoff_time)
    
    agent.restart_agent_learning()  # Reinicializa el conocimiento del agente

    # Se realiza la ejecución del agente
    avg_steps_per_episode = agent.run(algorithm=algorithm, policy=policy, penalty=penalty)
    
    return agent
```
