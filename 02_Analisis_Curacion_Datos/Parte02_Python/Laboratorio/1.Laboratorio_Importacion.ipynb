### 1.3 Despersonalizamos los datos y los guardarmos en un nuevo archivo

```python
building_permits_without_duplicates = building_permits[~building_permits.index.duplicated()]
```

```python
display(building_permits_without_duplicates.shape)
```

Vamos despersonalizar las siguientes columnas:
* Street Number
* Street Name
* Zipcode
* Location

Ya que a nuestro parecer son las más relevantes a la hora de ocultar su información.

Las tecnicas que aplicaremos seran las siguientes:
* Bucket values based on custom size ranges (`Street Number`)
* Pseudonymization by replacing input value with cryptographic hash (`Street Name`)
* Mask with character (`Zipcode`)
* Replace with infoType (`Location`)


Para `Street Number` reemplazaremos los valores por `LOW`, `MEDIUM`, `HIGH`:
* Si el `Street Number` esta entre 0 y 3333 $\longrightarrow$ `LOW`
* Si el `Street Number` esta entre 3334 y 6666 $\longrightarrow$ `MEDIUM`
* Si el `Street Number` es mayor a 6666 $\longrightarrow$ `HIGH`

Para `Street Name` aplicamos MD5 sobre los nombres para obtener un valor hexadecimal.

Para `Zipcode` reemplazamos los numberos a partir del 3 digito por `*` con el fin de ocultar parte del codigo postal.

Para `Location` reemplazamos su par (latitud, longitud) por el nombre del tipo de dato que corresponde en el dataset: `String`.

```python
# Bucket values based on custom size ranges
def bucketing(street_number):
    if 0 <= street_number <= 3333:
        return "LOW"
    elif 3333 < street_number <= 6666:
        return "MEDIUM"
    else:
        return "HIGH"
```

```python
from hashlib import md5

# Pseudonymization by replacing input value with cryptographic hash
def hashit(val):
    return md5(val.encode('utf-8')).hexdigest()
```

```python
# Mask with character
def hide_zipcode(zipcode):
    try:
        zipcode = int(zipcode)
    except:
        return zipcode

    new_zipcode = ""
    for index, c in enumerate(str(zipcode)):
        if index >= 2:
            new_zipcode += "*"
            continue
        new_zipcode += c
    
    return new_zipcode
```

```python
# Replace with infoType
def replace_by_infotype(location):
    infotype_location = "String"
    return infotype_location
```

```python
building_permits_without_duplicates['Street Number'].apply(bucketing).head(10)
```

```python
building_permits_without_duplicates['Street Name'].apply(hashit).head(10)
```

```python
building_permits_without_duplicates["Zipcode"].apply(hide_zipcode).head(10)
```

```python
building_permits_without_duplicates["Location"].apply(replace_by_infotype).head(10)
```
